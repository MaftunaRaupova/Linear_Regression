```python
""" Import insurance.csv into your notebook (Source).
● Use the data in the relevant columns to determine how age affects
insurance costs:
○ Plot a scatter plot with age on the x-axis and charges on the y-axis.
○ Using linear_model.LinearRegression() from sklearn, fit a model to
your data, and make predictions on data.
○ Plot another scatter plot with the best-fit line."""
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
#importing data from insurance.csv
df = pd.read_csv("insurance.csv")
df.head()
```

```python
# code below takes the insurance cost from data and scatters around. 

x = df.iloc[:,:1].values
y = df.iloc[:,-1].values

plt.scatter(x,y,color = "r")
plt.xlabel("Age of the person")
plt.ylabel("Insurance charge")
plt.show()
```

```python
# importing LeanerRegression from sklearn. 

from sklearn.linear_model import LinearRegression
```

```python
# code below fits the model and shows the prediction for the insurance cost as age grows as a green line. 

insurance_cost = LinearRegression()
insurance_cost.fit(x,y)

cost_pred = insurance_cost.predict(x)

plt.scatter(x,y,color = "r")
plt.xlabel("Age of the person")
plt.ylabel("Insurance charge")
plt.plot(x,cost_pred,color = 'g')
plt.show()
```
